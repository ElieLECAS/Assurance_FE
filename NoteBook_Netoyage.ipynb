```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

-Importez le Dataset CSV

```python
dataset_path = "Dataset_Brief.csv"
df = pd.read_csv(dataset_path)
```

-Voir le % de Valeurs null pour chaques cols.

-Drop les valeurs nulls du Dataset.

```python
missing_rate = df.isna().sum()/df.shape[0]
missing_rate
df = df.dropna()
```

-Vérifié les duplicatas de données dans le dataset.

-Drop les duplicatas en ne gardant que la premiere itération.

```python
print(df.duplicated().shape[0])

duplicates = df.duplicated()
duplicates_data = df[duplicates]


print("\nLignes avec des duplicatas:")
print(duplicates_data)

df = df.drop_duplicates(subset=["age", "sex", "bmi", "children", "smoker", "region"], keep='first')

df.shape[0]
```

-Variable pour encoder les cols objet.

-forcer valeurs des cols bmi et charges à passer en float.

-forcer valeurs de col children à passer en int.

-Creation des sous cols pour bmi & age en int.

-Pour chaque valeur unique dans col children créer une nouvelle col value.name en int.


```python
columns_to_encode = ['sex', 'smoker', 'region']

df['children'] = df['children'].astype(int)
df['bmi'] = df['bmi'].astype(float)
df['charges'] = df['charges'].astype(float)

df['Insuffisance pondérale'] = (df['bmi'] < 18.5).astype(int)
df['Poids_normal'] = ((df['bmi'] >= 18.5) & (df['bmi'] < 24.9)).astype(int)
df['Surpoids'] = ((df['bmi'] >= 24.9) & (df['bmi'] < 29.9)).astype(int)
df['Obésité_de_classe_I_(modérée)'] = ((df['bmi'] >= 29.9) & (df['bmi'] < 34.9)).astype(int)
df['Obésité_de_classe_II_(sévère)'] = (df['bmi'] >= 34.9).astype(int)

df['Jeune'] = (df['age'] < 18).astype(int)
df['Adulte'] = ((df['age'] >= 18) & (df['age'] < 35)).astype(int)
df['Adulte_moyen'] = ((df['age'] >= 35) & (df['age'] < 50)).astype(int)
df['Senior'] = ((df['age'] >= 50) & (df['age'] < 65)).astype(int)
df['Très_senior'] = (df['age'] >= 65).astype(int)

unique_values = df['children'].unique()

for value in unique_values:
    new_col_name = f'children_{value}'
    df[new_col_name] = (df['children'] == value).astype(int)	

df.describe()
```

-pour chaque value dans la variable encoder.
     - enregistre les variables unique des cols.

-pour chaque value dans variable unique.
     -créer une nouvel col avec pour nom is_value en int.


-rename is_yes & is_no en smoker & not_smoker.

-reprendre les cols sex & smoker en int 1 == female & yes.

-redéfini l'ordre des col du dataset .

```python

for col in columns_to_encode:
    if col not in df.columns:
        print(f"Warning: Column '{col}' not found in the DataFrame.")
        columns_to_encode.remove(col)

for col in columns_to_encode:
    unique_values = df[col].unique()
    for value in unique_values:
        new_col_name = f'is_{value.lower().replace(" ", "_")}'
        df[new_col_name] = (df[col] == value).astype(int)


df['sex'] = (df['sex'] == 'female').astype(int)
df['smoker'] = (df['smoker'] == 'yes').astype(int)

df.describe()
df = df.rename(columns={'is_yes': 'is_smoker', 'is_no': 'is_not_smoker'})

df = df[['age', 'Jeune', 'Adulte', 'Adulte_moyen', 'Senior', 'Très_senior', 'sex', 'is_female', 'is_male', 'bmi', 'Insuffisance pondérale', 'Poids_normal', 'Surpoids', 'Obésité_de_classe_I_(modérée)', 'Obésité_de_classe_II_(sévère)', 'children', 'children_0', 'children_1', 'children_2', 'children_3', 'children_4', 'children_5', 'smoker', 'is_smoker', 'is_not_smoker', 'is_southwest', 'is_southeast', 'is_northwest', 'is_northeast', 'charges']]

```
